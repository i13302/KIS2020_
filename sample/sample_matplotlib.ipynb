```python
# import library
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt

# Store the imported dataset in iris_dataset
iris = datasets.load_iris()
```

```python
# show line graph
plt.plot(iris.data)
plt.show()
```

```python
# show scatter plot
plt.plot(iris.data, 'o')
plt.show()
```

```python
iris_dataframe = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_dataframe["Name"] = pd.Series(iris.target)
iris_dataframe["Name"] = iris_dataframe["Name"].apply(lambda x: iris.target_names[x])
```

```python
plt.hist(iris_dataframe["petal length (cm)"], bins=20, color="green")
plt.xlabel("petal length")
plt.ylabel("counts")
plt.title("histgram of petal length")
```
